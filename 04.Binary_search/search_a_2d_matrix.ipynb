# Search a 2d Matirix

[Problem Link](https://leetcode.com/problems/search-a-2d-matrix/description/)



## Approach

- First check whether the target lies in the range of the matrix.
- If it does, then we can use binary search to find the target.
- First have to check in which row the target element may lie through binary search.
- Once we have the row, we can use binary search in that row to find the target element.

```python
def searchMatrix(matrix,target):
    if target>matrix[-1][-1] or target<matrix[0][0]:
        return False
    top=0
    bottum = len(matrix)-1

    while top<=bottum:
        mid = (top+bottum)//2
        if matrix[mid][0]>target:
            bottum=mid-1
        elif matrix[mid][-1]<target:
            top=mid+1
        else:
            break
    
    left = 0
    right = len(matrix[0])-1

    while left<=right:
        center = (left+right)//2
        if matrix[mid][center]>target:
            right=center-1
        elif matrix[mid][center]<target:
            left=center+1
        else:
            return True

    return False
```

```python
searchMatrix([[1,3,5,7],[10,11,16,20],[23,30,34,60]],3)
```
