# Binary Search

[Problem](https://leetcode.com/problems/binary-search/)



```python
def search(nums,target):
    left = 0
    right = len(nums) - 1
    while left <= right:
        mid = (left + right) // 2
        if nums[mid] < target:
            left = mid + 1
        elif nums[mid]>target:
            right=mid-1
        else:
            return mid
    return -1
```

```python
search([-1,0,3,5,9,12],9)
```

```python
search([-1,0,3,5,9,12],2)
```
