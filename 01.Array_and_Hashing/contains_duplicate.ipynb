# Contains Duplicate

[Problem Link](https://leetcode.com/problems/contains-duplicate/description/)




## Approach

- Create a set x 
- Iterate through the given array
- Check if the element exists in x
- If exists return true
- Add each element of the set to x

```python

def containsDuplicate( nums):
    x = set()
    for num in nums:
        if num not in x:
            x.add(num)
        else:
            return True
    return False
```

### testcase

```python
print(containsDuplicate([1,2,3,1]))
```
