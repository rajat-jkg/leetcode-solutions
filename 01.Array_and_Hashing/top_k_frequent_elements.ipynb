# Top k frequent elements



[Problem](https://leetcode.com/problems/top-k-frequent-elements/description/)

## Approach

- Create a hashmap
- Store the count of each element in the hashmap
- Convert the the hashmap into the list and sort on the basis of the count in reverse order
- Return the first k elements from the sorted tiems

```python
def topKFrequent(lst, k):
    count=dict()
    for ele in lst:
        if ele in count:
            count[ele] += 1
        else:
            count[ele] = 1
    return [x[0] for x in ((sorted(count.items() , key=lambda x:x[1], reverse=True))[:k])]
```

### Testcases

```python
print(topKFrequent([1,1,1,2,2,3],2))
```

```python
print(topKFrequent([0,1,3,0,0,5,0,0,0,9,6,78,9,9,1],3))
```
