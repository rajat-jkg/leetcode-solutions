# Gruop anagrams

[Problem Link](https://leetcode.com/problems/group-anagrams/description/)



## Approach

- Create an integer list of length 26 and an empty dictionary
- The element in the i-1<sup>th</sup> index represents the count of i<sup>th</sup> alphabet in(Example element in the 0th index repsents count of a)
- Iterate through the list of strings
- Iterate through the string and for each character, increment the corresponding count in the list
- Use the list as a key to store a list of stirngs in the dictionary
- The anagrams will be grouped as they'll have the same key

```python
def groupAnagrams(strs):
    groups = dict()
    for s in strs:
        charCount = [0]*26
        for c in s:
            charCount[ord(c)-97] += 1
        chars = tuple(charCount)
        if chars not in groups:
            groups[chars] = [s]
        else:
            groups[chars].append(s)
    return (groups.values())
```

### Testcases

```python
print(list(groupAnagrams(["eat","tea","tan","ate","nat","bat"])))
```

```python
print(list(groupAnagrams(['a'])))
```
