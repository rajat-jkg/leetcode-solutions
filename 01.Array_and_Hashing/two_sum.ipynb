# Tow sum



[Problem Link](https://leetcode.com/problems/two-sum/)

## Approach 

- Create a hashmap that maps list elements with its index
- Iterate through the list and for each element, check if its complement (target - element) exists
- If return the return the index and the complement

```python
def twoSum(nums, target):
    hash = dict()
    for i in range(len(nums)):
        hash[nums[i]]=i
    for i in range(len(nums)): 
        if target-nums[i] in hash:
            return [i,hash[target-nums[i]]]
```

### testcase

```python
print(twoSum([2,7,11,15],9))
```

```python
print(twoSum([3,2,4], 7))
```
